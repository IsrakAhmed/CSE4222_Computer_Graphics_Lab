```python
import matplotlib.pyplot as plt

def bresenham_line(xa, ya, xb, yb):
    dx = abs(xa - xb)
    dy = abs(ya - yb)
    p = 2 * dy - dx
    twoDy = 2 * dy
    twoDyDx = 2 * (dy - dx)

    x, y = xa, ya
    x_end = xb

    if xa > xb:
        x = xb
        y = yb
        x_end = xa

    plt.scatter(x, y, color='black')
    points = [[x, y]]  # Store the points for joining later
  
    while x < x_end:
        x += 1
        if p < 0:
            p += twoDy
        else:
            y += 1
            p += twoDyDx
        plt.scatter(x, y, color='black')
        points.append([x, y])

    # Join the points with a line
    x_coords, y_coords = zip(*points)
    plt.plot(x_coords, y_coords, color='blue')

    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title("Bresenham's Line")
    plt.grid(True)
    plt.show()
    print(points)

def main():
    xa, ya = map(int, input("Enter the starting point coordinates (x y): ").split())
    xb, yb = map(int, input("Enter the ending point coordinates (x y): ").split())

    bresenham_line(xa, ya, xb, yb)

main()
```

```python
import matplotlib.pyplot as plt

def bresenham_line(xa, ya, xb, yb):
    points = []

    dx = abs(xb - xa)
    dy = abs(yb - ya)

    x, y = xa, ya

    sx = 1 if xb > xa else -1
    sy = 1 if yb > ya else -1

    if dy <= dx:
        p = 2 * dy - dx
        for _ in range(dx + 1):
            points.append((x, y))
            x += sx
            if p < 0:
                p += 2 * dy
            else:
                y += sy
                p += 2 * (dy - dx)
    else:
        # Slope > 1
        p = 2 * dx - dy
        for _ in range(dy + 1):
            points.append((x, y))
            y += sy
            if p < 0:
                p += 2 * dx
            else:
                x += sx
                p += 2 * (dx - dy)

    # Plotting
    x_coords, y_coords = zip(*points)
    plt.scatter(x_coords, y_coords, color='black')
    plt.plot(x_coords, y_coords, color='blue')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title("Bresenham's Line")
    plt.grid(True)
    plt.gca().set_aspect('equal')
    plt.show()
    print(points)

def main():
    xa, ya = map(int, input("Enter the starting point coordinates (x y): ").split())
    xb, yb = map(int, input("Enter the ending point coordinates (x y): ").split())
    bresenham_line(xa, ya, xb, yb)

main()
```
