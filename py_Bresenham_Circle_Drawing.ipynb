```python
import matplotlib.pyplot as plt

# Function to plot all eight symmetric points of the circle
def drawCircle(xc, yc, x, y, points):
    points.extend([
        (xc + x, yc + y),
        (xc - x, yc + y),
        (xc + x, yc - y),
        (xc - x, yc - y),
        (xc + y, yc + x),
        (xc - y, yc + x),
        (xc + y, yc - x),
        (xc - y, yc - x)
    ])

# Bresenham Circle Drawing Algorithm
def circleBres(xc, yc, r):
    x = 0
    y = r
    d = 3 - 2 * r
    points = []
    drawCircle(xc, yc, x, y, points)

    while y >= x:
        if d > 0:
            y -= 1
            d = d + 4 * (x - y) + 10
        else:
            d = d + 4 * x + 6
        x += 1
        drawCircle(xc, yc, x, y, points)

    return points



# Main Program
xc, yc, r = 200, 200, 100
points = circleBres(xc, yc, r)

# Plot using matplotlib
x_vals, y_vals = zip(*points)
plt.figure(figsize=(6, 6))
plt.scatter(x_vals, y_vals, color='red', s=10)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Bresenham's Circle Drawing Algorithm")
plt.xlim(50, 350)
plt.ylim(50, 350)
plt.gca().invert_yaxis()  # match graphics.h coordinate system
plt.show()
```
