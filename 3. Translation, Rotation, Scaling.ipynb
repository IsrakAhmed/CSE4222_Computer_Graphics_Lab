```python
import matplotlib.pyplot as plt
import math

# Set screen dimensions (for consistent plot size)
WIDTH, HEIGHT = 800, 600

# Original triangle coordinates
x1, y1 = 300, 300
x2, y2 = 400, 100
x3, y3 = 500, 300

# Triangle points (closed loop for plotting)
original_triangle = [(x1, y1), (x2, y2), (x3, y3), (x1, y1)]

# ---------- 1. Translation ----------
tx, ty = 100, 50
tx1, ty1 = x1 + tx, y1 + ty
tx2, ty2 = x2 + tx, y2 + ty
tx3, ty3 = x3 + tx, y3 + ty
translated_triangle = [(tx1, ty1), (tx2, ty2), (tx3, ty3), (tx1, ty1)]

# ---------- 2. Rotation ----------
angle = 45  # degrees
rad = math.radians(angle)
rx1 = x1 * math.cos(rad) - y1 * math.sin(rad)
ry1 = x1 * math.sin(rad) + y1 * math.cos(rad)
rx2 = x2 * math.cos(rad) - y2 * math.sin(rad)
ry2 = x2 * math.sin(rad) + y2 * math.cos(rad)
rx3 = x3 * math.cos(rad) - y3 * math.sin(rad)
ry3 = x3 * math.sin(rad) + y3 * math.cos(rad)
rotated_triangle = [(rx1, ry1), (rx2, ry2), (rx3, ry3), (rx1, ry1)]

# ---------- 3. Scaling ----------
sx, sy = 1.5, 1.5
cx = (x1 + x2 + x3) / 3
cy = (y1 + y2 + y3) / 3
sx1 = cx + (x1 - cx) * sx
sy1 = cy + (y1 - cy) * sy
sx2 = cx + (x2 - cx) * sx
sy2 = cy + (y2 - cy) * sy
sx3 = cx + (x3 - cx) * sx
sy3 = cy + (y3 - cy) * sy
scaled_triangle = [(sx1, sy1), (sx2, sy2), (sx3, sy3), (sx1, sy1)]

# ---------- Plotting ----------
plt.figure(figsize=(8, 6))
plt.title("2D Transformation of Triangle")

# Plot all triangles
def plot_triangle(tri, color, label):
    xs, ys = zip(*tri)
    plt.plot(xs, ys, color=color, label=label, linewidth=2)

plot_triangle(original_triangle, 'blue', 'Original')
plot_triangle(translated_triangle, 'red', 'Translated')
plot_triangle(rotated_triangle, 'green', 'Rotated')
plot_triangle(scaled_triangle, 'black', 'Scaled')

# Aesthetic adjustments
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
plt.xlim(0, WIDTH)
plt.ylim(0, HEIGHT)
plt.xlabel("X")
plt.ylabel("Y")
plt.gca().invert_yaxis()  # To match screen-style (origin at top-left like Pygame)
plt.show()
```
