# Draw the Snowflake Pattern with Fractal Geometry

```python
import matplotlib.pyplot as plt
import numpy as np

# Recursive Koch Snowflake line function
def snowflake(x1, y1, x5, y5, it):
    if it > 0:
        dx = (x5 - x1) / 3
        dy = (y5 - y1) / 3

        # Calculate intermediate points
        x2, y2 = x1 + dx, y1 + dy
        x3 = (x1 + x5) / 2 + np.sqrt(3) * (y1 - y5) / 6
        y3 = (y1 + y5) / 2 + np.sqrt(3) * (x5 - x1) / 6
        x4, y4 = x1 + 2 * dx, y1 + 2 * dy

        # Recursive calls
        snowflake(x1, y1, x2, y2, it - 1)
        snowflake(x2, y2, x3, y3, it - 1)
        snowflake(x3, y3, x4, y4, it - 1)
        snowflake(x4, y4, x5, y5, it - 1)

    else:
        # Base case: draw line
        plt.plot([x1, x5], [y1, y5], color='blue', lw=1)

# Setup and draw
plt.figure(figsize=(6, 6))
plt.axis('equal')
plt.axis('off')

# Triangle base points
points = [(250, 15), (50, 350), (450, 350)]
iteration = 4

for i in range(3):
    x1, y1 = points[i]
    x2, y2 = points[(i + 1) % 3]
    snowflake(x1, y1, x2, y2, iteration)

plt.title("Fractal Geometry - Koch Snowflake", fontsize=12)
plt.show()
```
