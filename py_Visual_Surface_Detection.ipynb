```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create figure and axes
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(0, 400)
ax.set_ylim(0, 400)
ax.set_aspect('equal')
ax.set_facecolor('white')

# Draw rectangles (later ones hide earlier ones)
# (x, y) = bottom-left corner; in matplotlib, y increases upward
rect1 = patches.Rectangle((100, 100), 100, 100, facecolor='red', edgecolor='black')
rect2 = patches.Rectangle((130, 130), 100, 100, facecolor='blue', edgecolor='black')
rect3 = patches.Rectangle((160, 160), 100, 100, facecolor='green', edgecolor='black')

# Add rectangles in order (simulate visibility)
ax.add_patch(rect1)  # drawn first (bottom layer)
ax.add_patch(rect2)  # drawn next (middle layer)
ax.add_patch(rect3)  # drawn last (top layer)

# Add title and show
plt.title("Hidden Surface Elimination Simulation")
plt.axis('off')  # hide axes
plt.show()
```
